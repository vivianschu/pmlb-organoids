## Add Batch Information

```python
manifest_final = pd.read_csv('/Users/vivianschu/repos/pmlb-organoids/metadata/260313_manifest_final.tsv', sep='\t')
manifest_final
```

```python
manifest_replicate = pd.read_csv('/Users/vivianschu/repos/pmlb-organoids/metadata/260225_manifest_replicates.csv')
manifest_replicate
```

```python
# final_sample_name to combine sample_name and group_name, keep group_name if it exists instead
def combine_sample_group(row):
    if pd.notna(row['group_name']):
        return row['group_name']
    else:
        return row['sample_name']

manifest_replicate['final_sample_name'] = manifest_replicate.apply(combine_sample_group, axis=1)
manifest_replicate
```

```python
# add column indicating the folder origin, after /cluster/projects/livingbank/Data/RNAseq/
def extract_folder(path):
    key = '/cluster/projects/livingbank/Data/RNAseq/'
    if key in path:
        parts = path.split(key)[1].split('/')
        if parts[0] == 'RNAseq_Niki_Novogene' and len(parts) > 1:
            return f"{parts[0]}/{parts[1]}"
        else:
            return parts[0]
    else:
        return None


manifest_replicate['batch'] = manifest_replicate['mate1'].apply(extract_folder)
manifest_replicate
```

```python
# map manifest_replicate['batch'] to manifest using 'final_sample_name' and 'sample'
batch_mapping = manifest_replicate.set_index('final_sample_name')['batch'].to_dict()
manifest_final['batch'] = manifest_final['sample'].map(batch_mapping)

# also add mate1 and mate2 for historical
mate1_mapping = manifest_replicate.set_index('final_sample_name')['mate1'].to_dict()
mate2_mapping = manifest_replicate.set_index('final_sample_name')['mate2'].to_dict()
manifest_final['mate1'] = manifest_final['sample'].map(mate1_mapping)
manifest_final['mate2'] = manifest_final['sample'].map(mate2_mapping)
manifest_final
```

```python
# which manifest['batch'] are NaN?
nan_batches = manifest_final[manifest_final['batch'].isna()]['sample']
print(nan_batches)
print(len(nan_batches))
```

```python
# export csv as sample_manifest.tsv
manifest_final.to_csv('/Users/vivianschu/repos/pmlb-organoids/metadata/260313_manifest_batch.tsv', sep='\t', index=False)
```

## Clean Metadata

```python
# load gene tpm data
gene_tpm = pd.read_csv('/Users/vivianschu/repos/pmlb-organoids/data/combined_TPM.txt', sep='\t', index_col=0)
print(gene_tpm.shape)

# load star data
star_df = pd.read_csv('/Users/vivianschu/repos/pmlb-organoids/data/star_mapping_stats.tsv', sep='\t')
# load metadata
metadata_df = pd.read_csv('/Users/vivianschu/repos/pmlb-organoids/metadata/PMLB_LongTermTumor_Organoid_20260107.csv')
# remove unnamed col
metadata_df = metadata_df.loc[:, ~metadata_df.columns.str.contains('^Unnamed')]

# in new df, create col sample from gene_tpm columns except for first column
metadata = pd.DataFrame({'sample': gene_tpm.columns})
metadata.head()
```

```python
# map sample to PMLB_organoidID in metadata_df to get batch info
# if NaN, strip chars after '_' from sample name and try again
def map_to_organoid_id(sample_name):
    if sample_name in metadata_df['PMLB_organoidID'].values:
        return sample_name
    else:
        stripped_name = sample_name.split('_')[0]
        if stripped_name in metadata_df['PMLB_organoidID'].values:
            return stripped_name
        else:
            return None

metadata['PMLB_organoidID'] = metadata['sample'].apply(map_to_organoid_id)

metadata
```

```python
# add columns 'PMLB_organoidID', cancer_type, 'oncotree_code', 'primary_tumor_site', 'organoid_sample_class',  from metadata_df
metadata = metadata.merge(metadata_df[['PMLB_organoidID', 'cancer_type', 'oncotree_code', 'primary_tumor_site', 'organoid_sample_class']], left_on='PMLB_organoidID', right_on='PMLB_organoidID', how='left')

# remove the word ' Cancer' from cancer_type
metadata['cancer_type'] = metadata['cancer_type'].str.replace(' Cancer', '', regex=False)
metadata
```

```python
# combine manifest['batch'] and metadata
metadata = metadata.merge(manifest_final[['sample', 'batch']], left_on='sample', right_on='sample', how='left')

# export as metadata.to_csv
metadata.to_csv('/Users/vivianschu/repos/pmlb-organoids/metadata/metadata.csv', index=False)
metadata.head()
```
